```python
import pandas as pd

df = pd.read_csv('spam.csv', encoding='latin-1')
print(df.shape)
print(df.head())
```

```python
# Sadece işimize yarayan 2 sütunu alalım
df = df[['v1', 'v2']]
df.columns = ['label', 'message']
print(df.head())
print(df['label'].value_counts())
```

```python
import matplotlib.pyplot as plt

# Spam ve ham dağılımı grafiği
df['label'].value_counts().plot(kind='bar', color=['green', 'red'])
plt.title('Spam vs Ham Dağılımı')
plt.xlabel('Etiket')
plt.ylabel('Sayı')
plt.xticks(rotation=0)
plt.show()
```

```python
# Mesaj uzunluğu analizi
df['length'] = df['message'].apply(len)

df['length'].hist(by=df['label'], bins=50, figsize=(12,4), color='steelblue')
plt.suptitle('Mesaj Uzunluğu Dağılımı')
plt.show()

print(df.groupby('label')['length'].mean())
```

```python
# Eksik veri kontrolü
print("Eksik değerler:")
print(df.isnull().sum())

# Veri temizleme - metin küçük harfe çevir, noktalama sil
import re
import string

def temizle(metin):
    metin = metin.lower()
    metin = re.sub(r'\d+', '', metin)
    metin = metin.translate(str.maketrans('', '', string.punctuation))
    metin = metin.strip()
    return metin

df['clean_message'] = df['message'].apply(temizle)
print("\nTemizlenmiş örnek mesajlar:")
print(df[['message', 'clean_message']].head(3))
```

```python
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.naive_bayes import MultinomialNB
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, classification_report

# Label'ı sayıya çevir (spam=1, ham=0)
df['label_num'] = df['label'].map({'spam': 1, 'ham': 0})

# Train/test split
X_train, X_test, y_train, y_test = train_test_split(
    df['clean_message'], df['label_num'], test_size=0.2, random_state=42)

# TF-IDF
tfidf = TfidfVectorizer(max_features=3000)
X_train_tfidf = tfidf.fit_transform(X_train)
X_test_tfidf = tfidf.transform(X_test)

print("Eğitim seti:", X_train_tfidf.shape)
print("Test seti:", X_test_tfidf.shape)
```

```python
# MODEL 1: Naive Bayes
nb_model = MultinomialNB()
nb_model.fit(X_train_tfidf, y_train)
nb_pred = nb_model.predict(X_test_tfidf)

print("=== Naive Bayes Sonuçları ===")
print("Accuracy:", accuracy_score(y_test, nb_pred))
print(classification_report(y_test, nb_pred, target_names=['Ham', 'Spam']))

# MODEL 2: Lojistik Regresyon
lr_model = LogisticRegression(max_iter=1000)
lr_model.fit(X_train_tfidf, y_train)
lr_pred = lr_model.predict(X_test_tfidf)

print("=== Lojistik Regresyon Sonuçları ===")
print("Accuracy:", accuracy_score(y_test, lr_pred))
print(classification_report(y_test, lr_pred, target_names=['Ham', 'Spam']))
```

```python
import seaborn as sns
from sklearn.metrics import confusion_matrix

fig, axes = plt.subplots(1, 2, figsize=(12, 4))

for ax, pred, title in zip(axes, [nb_pred, lr_pred], 
                            ['Naive Bayes', 'Lojistik Regresyon']):
    cm = confusion_matrix(y_test, pred)
    sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', ax=ax,
                xticklabels=['Ham', 'Spam'],
                yticklabels=['Ham', 'Spam'])
    ax.set_title(title)
    ax.set_ylabel('Gerçek')
    ax.set_xlabel('Tahmin')

plt.tight_layout()
plt.show()
```
